## Module 2: Data Cleaning and Preparation¶

```python
# Sample dataset with messy data
customer_data = {
    'Customer_ID': ['A001', 'A002', 'A001', 'A003', 'A004', 'A005'],
    'Name': ['John Smith', 'jane doe', 'John Smith', 'Robert Johnson', 'LISA BROWN', 'David Wilson'],
    'Email': ['john@example.com', 'jane@example.com', 'john@example.com', np.nan, 'lisa@example.com', 'david@example'],
    'Purchase_Date': ['2023-01-15', '01/20/2023', '2023-01-15', '02-15-2023', '2023/03/10', '2023-04-05'],
    'Amount': ['$100.50', 150, '$100.50', '200', 175.75, '$250.00']
}

customer_df = pd.DataFrame(customer_data)
print("\nOriginal customer data:")
customer_df
```

```python
# Identifying and removing duplicates
print("\nDuplicate rows check:")
print(customer_df.duplicated())

print("\nDataFrame after removing duplicates:")
customer_df_no_dupes = customer_df.drop_duplicates()
customer_df_no_dupes
```

```python
# Standardizing text data (names to title case)
customer_df_no_dupes.loc[:, 'Name'] = customer_df_no_dupes['Name'].str.title()
print("\nNames standardized to title case:")
print(customer_df_no_dupes['Name'])
```

```python
# Validating emails (basic check for @)
customer_df_no_dupes.loc[:, 'Valid_Email'] = customer_df_no_dupes['Email'].str.contains('@', na=False)
print("\nValid email check:")
print(customer_df_no_dupes[['Email', 'Valid_Email']])

```

```python
# Converting date strings to datetime objects
# First, create a copy to work with
clean_df = customer_df_no_dupes.copy()
```

```python
# Function to handle multiple date formats
def parse_dates(date_str):
    if pd.isna(date_str):
        return pd.NaT
    
    formats = ['%Y-%m-%d', '%m/%d/%Y', '%m-%d-%Y', '%Y/%m/%d']
    
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except:
            continue
    
    return pd.NaT

# Apply function to convert dates
clean_df['Purchase_Date'] = clean_df['Purchase_Date'].apply(parse_dates)
print("\nPurchase dates converted to datetime:")
print(clean_df[['Purchase_Date']])
```

```python
# Standardizing currency/numeric values
def clean_amount(amount):
    if isinstance(amount, (int, float)):
        return amount
    
    # Remove currency symbols and commas, then convert to float
    if isinstance(amount, str):
        return float(amount.replace('$', '').replace(',', ''))
    
    return np.nan

clean_df['Amount'] = clean_df['Amount'].apply(clean_amount)
print("\nStandardized amounts:")
print(clean_df[['Amount']])
```

### 2.2: Data Transformation Techniques

```python
# Sample dataset for transformation examples
sales_data = {
    'Date': pd.date_range(start='2023-01-01', periods=10),
    'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C'],
    'Region': ['East', 'East', 'West', 'East', 'West', 'West', 'East', 'West', 'East', 'West'],
    'Sales': [100, 200, 150, 120, 180, 160, 210, 190, 140, 230],
    'Units': [10, 15, 12, 8, 14, 13, 16, 15, 11, 18]
}

sales_df = pd.DataFrame(sales_data)
print("\nSales data:")
sales_df.head()
```

```python
# Pivot tables
# Sales by Product and Region
pivot_sales = sales_df.pivot_table(
    values='Sales',
    index='Product',
    columns='Region',
    aggfunc='sum'
)
print("\nPivot table - Sales by Product and Region:")
pivot_sales
```

```python
# Reshaping with melt (unpivot)
melted_sales = pivot_sales.reset_index().melt(
    id_vars='Product',
    value_name='Total_Sales',
    var_name='Region'
)
print("\nMelted (unpivoted) sales data:")
melted_sales
```

```python
# Grouping and aggregating
# Total sales and average units by product
product_summary = sales_df.groupby('Product').agg({
    'Sales': 'sum',
    'Units': 'mean'
}).rename(columns={'Sales': 'Total_Sales', 'Units': 'Avg_Units'})

print("\nSummary by product:")
product_summary
```

```python
# Multiple groupings
region_product_summary = sales_df.groupby(['Region', 'Product']).agg({
    'Sales': ['sum', 'mean'],
    'Units': ['sum', 'mean']
})
print("\nSummary by region and product:")
region_product_summary
```

```python
# Applying functions
# Add a column with the sales per unit
sales_df['Sales_per_Unit'] = sales_df['Sales'] / sales_df['Units']
print("\nSales data with sales per unit:")
sales_df

```

```python
# Using apply to run a function on each row
def sales_category(row):
    if row['Sales'] < 150:
        return 'Low'
    elif row['Sales'] < 200:
        return 'Medium'
    else:
        return 'High'

sales_df['Sales_Category'] = sales_df.apply(sales_category, axis=1)
print("\nSales data with categories:")
sales_df
```

```python
# Merging datasets
# Create a second dataset with product information
product_info = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D'],
    'Category': ['Electronics', 'Clothing', 'Home', 'Office'],
    'Price': [50, 30, 40, 25]
})
product_info
```

```python
# Merge sales data with product info
merged_data = pd.merge(
    sales_df, 
    product_info, 
    on='Product', 
    how='left'
)
print("\nMerged sales and product data:")
merged_data
```

#### Calculate rolling average (3-month trend)

```python
# Calculate rolling average (3-month trend)

monthly_sales['Sales_3M_Avg'] = monthly_sales['Sales'].rolling(window=3).mean()
print("\nSales with 3-month rolling average:")
monthly_sales[['Month', 'Sales', 'Sales_3M_Avg']]
```

##### How it works:
- For **January**, there's no 3-month window, so the rolling average cannot be calculated.
- For **February**, it also doesn't have a full 3-month window, so it's still `NaN` (Not a Number).
- Starting from **March**, a 3-month window is available. You take the average of January, February, and March sales to get the rolling average for March.
- For **April**, the average of February, March, and April sales is used to calculate the rolling average, and so on.

##### Why it's useful:
- **Trend identification**: It helps smooth out irregularities and shows the underlying trend in sales over a period of time.
- **Better forecasting**: By eliminating short-term fluctuations, you can make more accurate predictions about future sales.
- **Performance evaluation**: It can help businesses evaluate if their sales are consistently improving or declining over time.

For example, if you see a rolling average of sales consistently rising, it indicates a positive trend, while a rolling average that declines suggests a negative trend, even if individual months might show sporadic fluctuations.

                                                                                                                                          

#### Customer cohort analysis example
##### Sample customer transaction data



```python
import random
random.seed(42)  # For reproducibility


# Generate sample data
dates = pd.date_range(start='2023-01-01', end='2023-06-30')
customer_ids = [f'C{i:03d}' for i in range(1, 101)]

# Create transactions
transactions = []
for _ in range(1000):
    customer = random.choice(customer_ids)
    date = random.choice(dates)
    amount = round(random.uniform(10, 200), 2)
    transactions.append({
        'customer_id': customer,
        'date': date,
        'amount': amount
    })

# Create DataFrame
trans_df = pd.DataFrame(transactions)
trans_df.head()
```

```python
# Add month column for cohort analysis
trans_df['month'] = trans_df['date'].dt.to_period('M')
trans_df.head()
```

```python
# Identify first purchase month for each customer (cohort)
cohort_data = trans_df.groupby('customer_id')['month'].min().reset_index()
cohort_data.columns = ['customer_id', 'cohort_month']
cohort_data.head()
```

```python
# Merge with original data
trans_df = trans_df.merge(cohort_data, on='customer_id')
trans_df.head()
```

```python
# Calculate months since first purchase
trans_df['months_since_first'] = (trans_df['month'] - trans_df['cohort_month']).apply(lambda x: x.n)
trans_df.head()
```

```python
# Create cohort table (retention analysis)
cohort_table = trans_df.groupby(['cohort_month', 'months_since_first']).agg({
    'customer_id': pd.Series.nunique
}).reset_index()
cohort_table.head()
```

```python
# Pivot to create retention table
retention_table = cohort_table.pivot(
    index='cohort_month',
    columns='months_since_first',
    values='customer_id'
)
retention_table
```

```python
# Calculate retention rates
retention_rates = retention_table.div(retention_table[0], axis=0) * 100

print("\nCustomer retention rates by cohort (%):")
retention_rates.round(1)
```

### 3.2: Research and Statistical Analysis

```python
# Generate sample dataset for research
np.random.seed(42)
n_samples = 100

research_data = pd.DataFrame({
    'Group': np.random.choice(['A', 'B', 'C'], n_samples),
    'Age': np.random.normal(35, 10, n_samples).round().astype(int),
    'Treatment': np.random.choice(['Yes', 'No'], n_samples),
    'Score_Before': np.random.normal(50, 15, n_samples).round(),
    'Score_After': None  # Will fill in based on treatment effect
})

# Simulate treatment effect
for i, row in research_data.iterrows():
    base = row['Score_Before']
    if row['Treatment'] == 'Yes':
        # Treatment has positive effect with some noise
        effect = np.random.normal(15, 5)
    else:
        # Control group has smaller effect with some noise
        effect = np.random.normal(5, 3)
    
    research_data.at[i, 'Score_After'] = base + effect

print("\nResearch study data sample:")
research_data.head()
```

####  Descriptive statistics by group

```python
# Descriptive statistics by group
group_stats = research_data.groupby('Group').agg({
    'Age': ['mean', 'median', 'std'],
    'Score_Before': ['mean', 'median', 'std'],
    'Score_After': ['mean', 'median', 'std']
})
print("\nDescriptive statistics by group:")
group_stats
```

#### Calculate improvement

```python
# Calculate improvement
research_data['Improvement'] = research_data['Score_After'] - research_data['Score_Before']
research_data.head()
```

#### Compare treatment vs control

```python
# Compare treatment vs control
treatment_effect = research_data.groupby('Treatment')['Improvement'].agg(['mean', 'std', 'count'])
print("\nTreatment effect analysis:")
treatment_effect
```

#### Correlation analysis

```python
# Correlation analysis
correlation = research_data[['Age', 'Score_Before', 'Score_After', 'Improvement']].corr()
print("\nCorrelation matrix:")
correlation
```

#### Time series example

```python
#### Time series example
# Generate time series data
dates = pd.date_range(start='2020-01-01', periods=100, freq='D')
ts_data = pd.DataFrame({
    'Date': dates,
    'Value': np.cumsum(np.random.normal(0, 1, 100)) + 100  # Random walk with drift
})
ts_data.head()
```

```python
# Set date as index
ts_data.set_index('Date', inplace=True)
ts_data.head()
```

```python
# Calculate rolling statistics
ts_data['Rolling_Mean'] = ts_data['Value'].rolling(window=7).mean()
ts_data['Rolling_Std'] = ts_data['Value'].rolling(window=7).std()

print("\nTime series data with rolling statistics:")
ts_data.head(10)
```

```python
# Resampling time series (e.g., to weekly)
weekly_data = ts_data.resample('W').agg({
    'Value': 'mean',
    'Rolling_Mean': 'mean',
    'Rolling_Std': 'mean'
})
print("\nWeekly resampled data:")
weekly_data.head()
```

### 4.1 Personal Finance Project

```python
# Sample personal finance data
dates = pd.date_range(start='2023-01-01', periods=100)
categories = ['Groceries', 'Dining', 'Transport', 'Shopping', 'Entertainment', 'Utilities', 'Rent', 'Income']
weights = [0.15, 0.1, 0.08, 0.07, 0.05, 0.1, 0.25, 0.2]

# Generate transactions
transactions = []
for _ in range(300):
    date = random.choice(dates)
    category = random.choices(categories, weights=weights)[0]
    
    if category == 'Income':
        amount = round(random.uniform(1000, 3000), 2)
    elif category == 'Rent':
        amount = -round(random.uniform(900, 1200), 2)
    elif category == 'Utilities':
        amount = -round(random.uniform(100, 300), 2)
    else:
        amount = -round(random.uniform(10, 200), 2)
    
    description = f"{category} - Transaction"
    
    transactions.append({
        'Date': date,
        'Category': category,
        'Description': description,
        'Amount': amount
    })

# Create DataFrame
finance_df = pd.DataFrame(transactions)
finance_df.sort_values('Date', inplace=True)
print("\nPersonal finance transactions:")
finance_df.head()
```

```python
# Monthly summary
finance_df['Month'] = finance_df['Date'].dt.to_period('M')
monthly_summary = finance_df.groupby(['Month', 'Category'])['Amount'].sum().unstack().fillna(0)
print("\nMonthly spending by category:")
monthly_summary
```

```python
# Calculate balance over time
finance_df['Running_Balance'] = finance_df['Amount'].cumsum()
print("\nRunning balance over time:")
finance_df[['Date', 'Amount', 'Running_Balance']].head(10)
```

```python
# Budget analysis
# Define a monthly budget
monthly_budget = {
    'Groceries': -500,
    'Dining': -300,
    'Transport': -200,
    'Shopping': -200,
    'Entertainment': -150,
    'Utilities': -250,
    'Rent': -1100
}
```

```python
# Calculate actual vs budget
current_month = finance_df['Month'].max()
current_month_data = finance_df[finance_df['Month'] == current_month]
current_month_data.head() 
```

```python
budget_analysis = []
for category, budget_amount in monthly_budget.items():
    actual_amount = current_month_data[current_month_data['Category'] == category]['Amount'].sum()
    remaining = budget_amount - actual_amount
    percentage_used = (actual_amount / budget_amount) * 100 if budget_amount != 0 else 0
    
    budget_analysis.append({
        'Category': category,
        'Budget': budget_amount,
        'Actual': actual_amount,
        'Remaining': remaining,
        'Percentage_Used': percentage_used
    })

budget_df = pd.DataFrame(budget_analysis)
print("\nBudget analysis for current month:")
budget_df
```

```python
# Generate sample COVID data for different countries
import pandas as pd
import numpy as np

# Generate date range with the corrected frequency alias
dates = pd.date_range(start='2020-03-01', end='2020-12-31', freq='ME')

# List of African countries
countries = ['Nigeria', 'South Africa', 'Egypt', 'Kenya', 'Ethiopia', 'Tanzania', 'Ghana', 'Uganda']

covid_data = []
for country in countries:
    # Generate some plausible patterns
    if country in ['Nigeria', 'South Africa', 'Egypt']:
        base_cases = np.random.randint(5000, 10000)
        growth_factor = np.random.uniform(1.5, 2.0)
    else:
        base_cases = np.random.randint(1000, 5000)
        growth_factor = np.random.uniform(1.2, 1.5)

    for i, date in enumerate(dates):
        # Simulate waves with some randomness
        wave_factor = 1 + 0.5 * np.sin(i / 2)
        monthly_cases = int(base_cases * (growth_factor ** (i/3)) * wave_factor)
        monthly_cases = monthly_cases * (1 + np.random.uniform(-0.1, 0.1))  # Add noise

        # Different recovery rates
        if country in ['Kenya', 'Ethiopia', 'Tanzania']:
            recovery_rate = np.random.uniform(0.7, 0.9)
        else:
            recovery_rate = np.random.uniform(0.5, 0.75)

        monthly_recoveries = int(monthly_cases * recovery_rate)

        covid_data.append({
            'Date': date,
            'Country': country,
            'New_Cases': int(monthly_cases),
            'Recoveries': int(monthly_recoveries),
            'Active_Cases': int(monthly_cases - monthly_recoveries)
        })

# Create DataFrame
covid_df = pd.DataFrame(covid_data)
print("\nCOVID-19 data sample for African countries:")
covid_df.head()

```

```python
# Basic data exploration
covid_summary = covid_df.groupby('Country').agg({
    'New_Cases': 'sum',
    'Recoveries': 'sum'
})
covid_summary
```

```python
# Calculate recovery rate
covid_summary['Recovery_Rate'] = covid_summary['Recoveries'] / covid_summary['New_Cases'] * 100
covid_summary = covid_summary.sort_values('New_Cases', ascending=False)

print("\nCOVID-19 summary by country:")
covid_summary
```

```python
# Time series analysis by country
covid_time_series = covid_df.pivot_table(
    index='Date',
    columns='Country',
    values='New_Cases',
    aggfunc='sum'
)

print("\nMonthly cases by country:")
covid_time_series.head()
```

```python
# 1. Customer Segmentation using RFM Analysis
# RFM = Recency, Frequency, Monetary Value

# Set the last date of the dataset as the reference point
last_date = supermarket_df['Date'].max()

# Create a customer RFM summary
rfm_data = supermarket_df.groupby('Customer_ID').agg({
    'Date': lambda x: (last_date - x.max()).days,  # Recency (days since last purchase)
    'Store_ID': 'count',  # Frequency (count of transactions)
    'Total': 'sum'  # Monetary (total amount spent)
}).rename(columns={
    'Date': 'Recency',
    'Store_ID': 'Frequency',
    'Total': 'Monetary'
})
rfm_data.head()
```

```python
# Rank the RFM metrics
rfm_data['R_Rank'] = rfm_data['Recency'].rank(ascending=False)  # Lower recency is better
rfm_data['F_Rank'] = rfm_data['Frequency'].rank(ascending=True)
rfm_data['M_Rank'] = rfm_data['Monetary'].rank(ascending=True)

# Normalize the ranks to create scores from 1-5
rfm_data['R_Score'] = (rfm_data['R_Rank'] / rfm_data['R_Rank'].max() * 5).round() + 1
rfm_data['F_Score'] = (rfm_data['F_Rank'] / rfm_data['F_Rank'].max() * 5).round() + 1
rfm_data['M_Score'] = (rfm_data['M_Rank'] / rfm_data['M_Rank'].max() * 5).round() + 1

# Calculate RFM Score
rfm_data['RFM_Score'] = rfm_data['R_Score'] * 100 + rfm_data['F_Score'] * 10 + rfm_data['M_Score']

# Segment customers based on RFM Score
def segment_customer(score):
    r = int(score / 100)
    f = int((score % 100) / 10)
    m = score % 10
    
    if r >= 4 and f >= 4 and m >= 4:
        return 'Top Customers'
    elif r >= 3 and f >= 3 and m >= 3:
        return 'High-Value Customers'
    elif r >= 3 and f >= 1 and m >= 2:
        return 'Medium-Value Customers'
    elif r >= 2 and f >= 2 and m >= 2:
        return 'Potential Loyalists'
    elif r <= 2 and f <= 2 and m <= 2:
        return 'Lost Customers'
    else:
        return 'Others'

rfm_data['Segment'] = rfm_data['RFM_Score'].apply(segment_customer)

print("\nCustomer segmentation based on RFM analysis:")
rfm_data[['Recency', 'Frequency', 'Monetary', 'RFM_Score', 'Segment']].head(10)
```

```python
# Count of customers by segment
segment_counts = rfm_data['Segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'Count']
print("\nNumber of customers by segment:")
print(segment_counts)
```

```python
# 2. Product Affinity Analysis
# Calculate the lift ratio for product pairs (a measure of association)

# Get counts of individual products
# Assuming supermarket_df, transactions, and pair_counts are already defined

# Get counts of individual products
product_counts = supermarket_df['Product'].value_counts()
total_transactions = len(transactions)

# Calculate lift for the top product pairs
top_pairs = pair_counts.head(20).copy()

# Initialize the 'Lift' column with a float data type
top_pairs['Lift'] = 0
top_pairs['Lift'] = top_pairs['Lift'].astype(float)

for idx, row in top_pairs.iterrows():
    # Count of first product
    count_product_1 = product_counts[row['Product_1']]
    # Count of second product
    count_product_2 = product_counts[row['Product_2']]

    # Calculate expected frequency if the products were independent
    expected_frequency = (count_product_1 * count_product_2) / total_transactions

    # Calculate lift (with error handling)
    if expected_frequency > 0:
        lift = row['Frequency'] / expected_frequency
        top_pairs.at[idx, 'Lift'] = lift

print("\nProduct affinity analysis (lift ratio):")
print(top_pairs.sort_values('Lift', ascending=False))
```

```python
# 3. Time Series Decomposition
# Analyze daily sales patterns and decompose into trend, seasonal, and residual components
daily_sales = supermarket_df.groupby('Date')['Total'].sum().reset_index()
daily_sales.set_index('Date', inplace=True)

# Create a time series
ts = daily_sales['Total']

# Decompose the time series
from statsmodels.tsa.seasonal import seasonal_decompose

# Detect and fill missing dates
full_date_range = pd.date_range(start=ts.index.min(), end=ts.index.max(), freq='D')
ts = ts.reindex(full_date_range)
ts.interpolate(method='linear', inplace=True)

# Perform decomposition with period=7 for weekly seasonality
# Fixed (handle missing values properly):
# Fill missing values before decomposition
ts = ts.fillna(method='ffill').fillna(method='bfill')
decomposition = seasonal_decompose(ts, model='additive', period=7)

# Get decomposition components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Package results
decomp_df = pd.DataFrame({
    'Original': ts,
    'Trend': trend,
    'Seasonal': seasonal,
    'Residual': residual
})

print("\nTime series decomposition (first few rows):")
decomp_df.head()
```

```python
# 4. Price Elasticity Analysis
# Estimate how price changes affect sales volume


# Create price bins for analysis
supermarket_df['Price_Bin'] = pd.cut(
    supermarket_df['Price'],
    bins=[0, 5, 10, 15, 20, 100],
    labels=['0-5', '5-10', '10-15', '15-20', '20+']
)

# Analyze average quantity sold by price bin for each category
price_elasticity = supermarket_df.groupby(['Category', 'Price_Bin'], observed=True).agg({
    'Quantity': 'mean',
    'Price': 'mean',
    'Total': 'sum',
    'Product': 'count'
}).rename(columns={'Product': 'Transactions'}).reset_index()

print("\nPrice elasticity analysis:")
price_elasticity.head()



```

```python
# 5. Customer Lifetime Value (CLV) Calculation
# Calculate the future value of customers based on their purchase patterns

# Analyze customer purchase frequency and average spend
customer_value = supermarket_df.groupby('Customer_ID').agg({
    'Date': [
        'min',  # First purchase date
        'max',  # Last purchase date
        'count'  # Number of purchase days
    ],
    'Total': 'sum'  # Total spend
}).reset_index()

# Flatten hierarchical column names
customer_value.columns = ['Customer_ID', 'First_Purchase', 'Last_Purchase', 'Purchase_Count', 'Total_Spend']

# Calculate days as customer
customer_value['Days_as_Customer'] = (customer_value['Last_Purchase'] - customer_value['First_Purchase']).dt.days + 1
customer_value['Days_as_Customer'] = customer_value['Days_as_Customer'].clip(lower=1)  #


# Calculate average purchase frequency (purchases per week)
customer_value['Purchase_Frequency'] = customer_value['Purchase_Count'] / (customer_value['Days_as_Customer'] / 7)

# Calculate average order value
customer_value['Avg_Order_Value'] = customer_value['Total_Spend'] / customer_value['Purchase_Count']

# Calculate customer value per week
customer_value['Weekly_Value'] = customer_value['Purchase_Frequency'] * customer_value['Avg_Order_Value']

# Calculate Simple CLV (52 weeks = 1 year projection)
customer_value['CLV_1Year'] = customer_value['Weekly_Value'] * 52

print("\nCustomer Lifetime Value analysis:")
customer_value[['Customer_ID', 'Purchase_Count', 'Total_Spend', 'Avg_Order_Value', 'Weekly_Value', 'CLV_1Year']].head(10)
```

```python
# Summary statistics
clv_summary = customer_value['CLV_1Year'].describe()
print("\nCLV Summary Statistics:")
clv_summary
```

```python
# 6. Cohort Analysis
# Analyze customer retention and spending patterns based on when they first purchased

# Create cohort data
supermarket_df['First_Purchase'] = supermarket_df.groupby('Customer_ID')['Date'].transform('min')
supermarket_df['Cohort_Month'] = supermarket_df['First_Purchase'].dt.to_period('M')
supermarket_df['Purchase_Month'] = supermarket_df['Date'].dt.to_period('M')
supermarket_df['Cohort_Index'] = (supermarket_df['Purchase_Month'] - supermarket_df['Cohort_Month']).apply(lambda x: x.n)

# Cohort size (number of customers)
cohort_size = supermarket_df.groupby('Cohort_Month')['Customer_ID'].nunique().reset_index()
cohort_size.columns = ['Cohort_Month', 'Customer_Count']
print("\nCohort sizes:")
print(cohort_size)

# Customer retention by cohort
cohort_retention = supermarket_df.groupby(['Cohort_Month', 'Cohort_Index'])['Customer_ID'].nunique().reset_index()
cohort_retention = cohort_retention.pivot(index='Cohort_Month', columns='Cohort_Index', values='Customer_ID').fillna(0)

# Calculate retention rate
cohort_retention_rate = cohort_retention.div(cohort_retention[0], axis=0) * 100
print("\nCohort retention rates (%):")
cohort_retention_rate.round(1)
```
